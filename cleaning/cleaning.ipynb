```python
import pandas as pd
```

```python
df_fooding = pd.read_csv('2024-12-20_lefooding_paris10', sep=';')
df_michelin = pd.read_csv('2024-12-20_michelin_paris10', sep=';')
```

```python
df_fooding.head()
```

```python
df_fooding['source'] = 'lefooding'
```

```python
df_michelin.head()
```

```python
df_michelin['source'] = 'michelin'
```

```python
df_michelin['address'] = df_michelin['street'].str.cat(df_michelin[['postcode','city']], sep=", ")
```

```python
df_michelin.head()
```

```python
df_michelin['cuisine_type'] = df_michelin['cuisine_type'].str.replace('[','').str.replace(']','').str.replace("'",'')
df_michelin['tags'] = df_michelin['tags'].str.replace('[','').str.replace(']','').str.replace("'",'')
```

```python
df_michelin
```

```python
df_fooding['cuisine_type'] = df_fooding['food_type'].str.replace('[','').str.replace(']','').str.replace("'",'')
```

```python
df_fooding.head()
```

```python
df_fooding['cuisine_type'] = df_fooding['cuisine_type'].str.split(',')
```

```python
df_fooding.head()
```

```python
df_michelin2 = df_michelin[['name','address','price','cuisine_type','tags','source']]
```

```python
df_all = pd.concat([df_fooding, df_michelin2])
```

```python
df_all
```

```python
df_all
```

```python
df_all['name'] = df_all['name'].str.capitalize()
```

```python
df_all['name'].nunique()
```

```python
df_all['name'].value_counts()
```

```python
df_all.drop_duplicates(subset='name', keep='first', inplace=True)
```

```python
df_all['name'].value_counts()
```

```python
df_all['name'].unique()
```

```python
df_all['cuisine_type'] = df_all['cuisine_type'].str.split(',')
```

```python
df_all['tags'] = df_all['tags'].str.replace('[', '').str.replace(']', '').str.replace("'", '')
df_all['tags'] = df_all['tags'].str.split(',')
```

```python
df_all.head()
```

```python
df_all.price.unique()
```

```python
if 'Moins de 15 €' in df_all.price:
    df_all['clean_price'] = '€'
if 'Sans se priver' in df_all.price:
    df_all['clean_price'] = '€-€€'
if 'De 16 à 35 €' in df_all.price:
    df_all['clean_price'] = '€€'
if 'Se faire plaisir' in df_all.price:
    df_all['clean_price'] = '€€-€€€'
if 'De 36 à 50 €' in df_all.price:
    df_all['clean_price'] = '€€€'
if 'Pour une belle occasion' in df_all.price:
    df_all['clean_price'] = '€€€-€€€€'
if 'Plus de 51 €' in df_all.price:
    df_all['clean_price'] = '€€€€'
if 'Faire une folie' in df_all.price:
    df_all['clean_price'] = '€€€€€'
```

```python
df_all.head()
```

```python
df_all.to_csv('2024-12-20_all_paris10.csv', sep=';', index=False)
```
